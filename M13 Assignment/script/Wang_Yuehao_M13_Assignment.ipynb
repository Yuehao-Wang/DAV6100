### 2） Now briefly profile the dataset. Provide a response to the following:

- How many documents are in the dataset?

- What is the average size of the documents?

- What is the size of the database?

```python
status = db.command("dbstats")
pprint.pprint(status)
```

```python
# How many documents are in the dataset?
print("listingsAndReviews has %d document" %status['objects'])
```

```python
# What is the average size of the documents?
avg_size = status['avgObjSize'] / ( 1024 * 1024)
print("the average size of the documents is %0.4f MB" %avg_size)
```

```python
# What is the size of the database?
print("the size of the database is %0.2f MB" %(status['dataSize'] / (1024 * 1024)))
```

### 3) Now run queries to answer the following:

- How many listings were reviewed on 2016-01-31 in the listingsAndReviews collection?

- What is the property_type of the _id =’10084023’?

- How many listings have a property_type = ‘house’?

```python
# How many listings were reviewed on 2016-01-31 in the listingsAndReviews collection?
start_date = datetime.datetime(2016, 1, 30, 23, 59, 59)

end_date = datetime.datetime(2016, 1, 31, 23, 59, 59)

# count_reviewed = col.find( { "last_review": { "$gt": start_date, "$lte": end_date  } } ).count()
count_reviewed = col.count_documents( { "last_review": { "$gt": start_date, "$lte": end_date  } } )
print("There is %d that were reviewed on 2016-01-31 in the listingsAndReviews collection" %count_reviewed)
```

```python
# What is the property_type of the _id =’10084023’?
#query specific listing and store as result
query = {"_id":"10084023"}
result = col.find(query)

#print results
for x in result:
    print("the property_type of the _id =’10084023’ is " + x["property_type"])
```

```python
# How many listings have a property_type = ‘house’?
query = {"property_type":"House"}
house_count = col.count_documents(query)
print("the count of property_type = ‘House’ is %d" %house_count)
```

```python
# To check the above query for How many listings have a property_type = ‘house’?
query = {"property_type": re.compile("House", re.IGNORECASE)}
result = col.find(query, {"property_type" : 1})
for x in result:
    print(x)
```

The above prints show that there are many types containing house. Like, 'Earth house', 'Guesthouse', 'House', 'Townhouse' and 'Treehouse'. We should use {"property_type":"House"} to query result because Mongodb is case-sensitive.
